```python
#2. Create a table that displays the average height of players at each position.
'''
SELECT ASC(height), position FROM benn.college_football_players
WHERE height != 0
GROUP BY position
ORDER BY avg DESC
SELECT * FROM benn.college_football_players
'''
```
